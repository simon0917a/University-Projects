```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 定義欄位名稱並讀取資料 [cite: 20, 21, 22]
names = ['place', 'age', 'NoMale', 'NoFemale', 'DNoMale', 'DNoFemale', 'DMale_P', 'DFemale_P']
dataset = pd.read_csv('data for Lab2.csv', names=names)
```

```python
# Assuming dataset is already loaded via pandas as per Section 2
m1_female = np.polyfit(dataset['age'], dataset['DFemale_P'], 1)
b_female = m1_female[0] # Slope
a_female = m1_female[1] # Intercept
print(f"Linear estimation parameters: a = {a_female:.6f}, b = {b_female:.6f}")
```

```python
# The setting "2" indicates that a quadratic equation is used
m2_female = np.polyfit(dataset['age'], dataset['DFemale_P'], 2)
c_quad = m2_female[0] # Coefficient of x^2
b_quad = m2_female[1] # Coefficient of x
a_quad = m2_female[2] # Intercept
print(f"Quadratic parameters: a = {a_quad:.6f}, b = {b_quad:.6f}, c = {c_quad:.6f}")
```

```python
# Obtain predicted values for both models
pred_linear_f = np.polyval(m1_female, dataset['age'])
pred_quad_f = np.polyval(m2_female, dataset['age'])

# Calculate the square sum differences
E_linear_f = ((dataset['DFemale_P'] - pred_linear_f) ** 2).sum()
E_quad_f = ((dataset['DFemale_P'] - pred_quad_f) ** 2).sum()

print(f"Square sum difference (Straight-line): {E_linear_f:.6f}")
print(f"Square sum difference (Quadratic): {E_quad_f:.6f}")
```
